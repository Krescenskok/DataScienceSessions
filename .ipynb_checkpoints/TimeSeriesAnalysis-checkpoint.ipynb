## Time Series Analysis

A sequence of data points that occur in equal time intervals over some period of time. 

**Dependencies:**
- **Stationarity:** Statistical properties (mean, variance, autocorrelation) don't change over time.
    - Ex: Weather temperatures is not stationary due to having seasonal patterns

<br />

- **Autocorrelation:** How correlated a time series is with itself over time
    - **High Correlation:** Indicates that the data is related to previous days (weather temperatures)
    - **Low Correlation:** Indicates that there is no relationship between data points
    - **Negative Correlation:** Indicates a trend such as high stock prices followed by low stock prices (vice versa)
    
<img src="autocorrelation.jpg" width="500"/>



- **Missingness:** The dataset should be free of missing data and should have no missing time interval

## Stationary
- Based on the graph below, we can see that the data is not stationary since we see an upwards trend and the mean is chaning over time.

```python
#Plot the time series
df.plot(xlabel='Day', ylabel='Revenue', grid=True, 
        title='Revenue in Millions for the Last Two Years', 
        legend=False)
plt.show()
```

#### We can verify that the data is not stationary by using the Augmented Dickey-Fuller test
- A statistical significance test
- **Null Hypothesis:** Data is stationary
- **Alternate Hypothesis:** Data is not stationary

```python
from statsmodels.tsa.stattools import adfuller #for augmented dickey-fuller (d stationarity)

#perform test
adfresult = adfuller(df['Revenue'])
```

#### ADF Result Meaning:
Test Statistic

p-value

Number of lags

Number of observations

Critical values

```python
adfresult
```

#### Test Statistic:

```python
adfresult[0]   
```

#### p-value:
Since the p-value is not significant, we reject the null hypothesis, and conclude that the data is not stationary

```python
#p value 
adfresult[1] 
```

#### Critical Values:
The test statistic is greater than the critical values, so again, we can conclude that the data is not stationary

```python
#critical values
adfresult[4]
```

### Since the data is not stationary, the next step is to transform the data
- Take the differences between each point, take the logarithm of each point, etc...

```python
 # change actuals to differences
diff_df = df.diff()
```

##### We expect the first row to be Null since there is no data point for the first record to be compared to

```python
#check for Na on first row
diff_df.head()
```

##### Since the first record is null, we will drop it from our dataframe

```python
#drop na on first row
diff_df = diff_df.dropna() 
```

```python
diff_df.head()
```

##### Now we can see that the data is stationary and the mean/variance are the same throughout the dates

```python
#Plot the time series
diff_df.plot(grid=True)
plt.show()
```

```python
adfresult = adfuller(diff_df['Revenue'])
adfresult
```

#### Since the p-value is less than .05, it is significant so we fail to reject the null hypothesis and conclude that the data is stationary

```python
adfresult[1] 
```

## Auto ARIMA
- Works to find the optimal order of p,d, and q by identifying the lowest AIC/BIC metric values
- Hyperparameter tuning
- Performs a stepwise search to find the best model, searching through various combinations of p,d, and q parameters of ARIMA, and the P, D, and Q parameters of seasonal ARIMA (SARIMA) models

##### First Parenthesis:
- p: Number of autoregressive terms
- d: Number of nonseasonal differences
- q: Number of lagged forecast erros in the prediction equation (moving average)


##### Second Parenthesis: Seasonality

##### Third Parenthesis: Number of intervals representing the seasonality repeat

##### Model Metric: The lower the Akaike information criterion (AIC), the better the model

```python
from pmdarima import auto_arima #alternative to finding p/q

#trace = True prints out each iteration and the result of the model
stepwise_fit = auto_arima(df['Revenue'], trace=True, suppress_warnings=True)
```

#### Best model according to the auto_arima function suggests that the current value of the time series can be predicted based on its previous value after taking a first difference to remove any trend or seasonality
##### (1,1,0):
- 1 represents the AR order, indicating that the current value of the time series is linearly dependent on its previous value
- 1 represents the differencing order, indicating that the time series needs to be differenced once to make it stationary (removing any trend or seasonality)
- 0 represents the moving average, indicating that the model does not incorporate any moving average terms
##### (0,0,0):
- 0 represents the seasonal autoregressive (SAR) order, indicating that the model does not consider any seasonal autoregressive terms
- 0 represents the seasonal differencing order, indicating that the model does not include any seasonal differencing
- 0 represents the seasonal moving average, indicating the model does not include any seasonal moving average terms
##### [0]:
- Denotes that the model is not applied to any seaosnal component of the time series

```python
stepwise_fit.summary()
```

### ACF and PACF Plots

- The partial autocorrelation function shows strong correlation at lag one
- Since the first lag is outside of the confidence interval, it suggests that the AR (p) value in the ARIMA model is 1
- Since the difference was only taken once, d = 1 in the model
- There is no exponential decay on the PACF, so there is no moving average component, setting q = 0

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf #for plotting ACF/PACF (q/p of model)

# create figure to Plot ACF and PACF as subplots with 2 rows and 1 column
fig, axes = plt.subplots(2,1)

#Plot ACF of the differenced series
plot_acf(diff_df, lags=20, ax=axes[0])  

#Plot PACF of the differenced series
plot_pacf(diff_df, lags=20, ax=axes[1]) 
plt.show()
```

### Comparing Models
- Model 1: Results from auto arima

### Metrics:
- Akaike information criterion (AIC)
- Bayesian information criterion (BIC): Penalizes models with a large number of parameters

### Model 1

```python
from statsmodels.tsa.arima.model import ARIMA #for estimating an ARIMA model

# hyper parameter model from auto arima
model1 = ARIMA(train, order=(1,1,0))
results1 = model1.fit()

results1.summary()
```

```python
results1.aic
```

```python
results1.bic
```

### Model 2

```python
import statsmodels.api as sm #for sarimax

# seasonal model from decomposition - 7 day seasonality
model2=sm.tsa.statespace.SARIMAX(train['Revenue'],order=(1, 1, 0),seasonal_order=(1,1,0,7))
results2=model2.fit()

results2.summary()
```

```python
results2.aic #933.929
```

```python
results2.bic
```

### Model 3

```python
# ARMA(1,1) model - no seasonality
model3 = ARIMA(train, order=(1,0,1))
results3 = model3.fit()

# See ARMA(1,1) results
results3.summary()
```

```python
results3.aic
```

```python
results3.bic
```

### Comparing Metrics
- Model 1: AIC value of 727.65, BIC value of 736.23
- Model 2: AIC value of 933.9, BIC value of 996.90
- Model 3: AIC value of 764.65, BIC value of 781.86

Since the lower AIC and BIC values are better, the auto arima gave the best results by not using weakly seasonality 

### Perform model on test set

```python
#-------------Build model on train and compare test to prediction

#build training model
#by setting trend='t', we are assuming that the time series has a trend that can be described by a linear function
model = ARIMA(train, trend='t',order=(1,1,0))
results = model.fit()
```

```python
#predict against test with train
#predict the same months as test - last 25% since there was a 75/25 split
prediction = results.predict(start='2023-06-30', end='2023-12-31')
prediction
```

```python
#plot predict vs test
prediction.plot(legend=True, title = 'Prediction vs Test')
test['Revenue'].plot(legend=True)
```

#### Comparing prediction to test:
- The difference in the prediction results and the test set is measured using the root mean squared error as a metric
    -  Formula: $(\sqrt{Σ(y_{true} - y_{pred})^2})/n$
- RSME: The smaller the value is compared to the mean means that the model's predictions are closer to the actual values
- The rsme of 1.99 compared to the mean revenue of 13.54 shows a low margin of error and a very good fit

```python
test['Revenue'].mean()
```

```python
from sklearn.metrics import mean_squared_error #for arima metric
from math import sqrt

rmse = sqrt(mean_squared_error(prediction,test))
rmse 
```
